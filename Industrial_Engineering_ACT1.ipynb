```python
# Constantes de los resortes (N/m)
k1 = 100
k2 = 50
k3 = 80
k4 = 200

# Aceleración de la gravedad
g = 9.81

# Casos de fuerza
F1 = 1500              # Caso A: fuerza dada directamente en Newtons
F2 = 1500 * g          # Caso B: 1500 kgf convertidos a Newtons

# Función que calcula los x dado un valor de F
def calcular_x(F, k1, k2, k3, k4):
    x1 = F / k1
    x2 = x1 + F / k2
    x3 = x2 + F / k3
    x4 = x3 + F / k4
    return x1, x2, x3, x4

# Cálculos
x1_caseA, x2_caseA, x3_caseA, x4_caseA = calcular_x(F1, k1, k2, k3, k4)
x1_caseB, x2_caseB, x3_caseB, x4_caseB = calcular_x(F2, k1, k2, k3, k4)

# Mostrar resultados
print("=== Caso A: F = 1500 N ===")
print(f"x1 = {x1_caseA:.3f} m")
print(f"x2 = {x2_caseA:.3f} m")
print(f"x3 = {x3_caseA:.3f} m")
print(f"x4 = {x4_caseA:.3f} m")

print("\n=== Caso B: F = 1500 kgf (≈ 14715 N) ===")
print(f"x1 = {x1_caseB:.3f} m")
print(f"x2 = {x2_caseB:.3f} m")
print(f"x3 = {x3_caseB:.3f} m")
print(f"x4 = {x4_caseB:.3f} m")
```
